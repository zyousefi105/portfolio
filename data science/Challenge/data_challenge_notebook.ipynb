```python
from sklearn.metrics import confusion_matrix
#Profit calculator
def profit_calculator(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    tp = cm[1,1]
    fp = cm[0,1]
    return 100*tp - 1000*fp

profit_calculator(y_test, predictions)
```

```python
from sklearn.metrics import f1_score

def performance_metric(y_true, y_predict):
  
    score = f1_score(y_true,y_predict)   
    return score
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import ShuffleSplit

def fit_model(X, y):
   
    cv_sets = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.20, random_state = 0)
    classifier = DecisionTreeClassifier()
    params = {'max_depth':list(range(1,11))}
    scoring_fnc = make_scorer(performance_metric)
    grid = GridSearchCV(classifier, params, scoring=scoring_fnc, cv=cv_sets)
    grid = grid.fit(X, y)

    return grid
```

```python
gs = fit_model(X_train, y_train)

print gs.best_score_ 
print gs.best_estimator_ 
```

```python
predict = gs.best_estimator_.predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
def predictor(model, X_train, X_test, y_train, thres=.5):
    #Initialize Model
    mod = model()
    #Train model on training data
    mod.fit(X_train, y_train)
    #Make predictions on the testing set (X_test)
    preds = mod.predict_proba(X_test)
    #preds >= thres returns boolean values
    #np.where replaces True with 1 and False with 0
    threshold_adjusted_preds = np.where(preds>=thres, 1, 0)
    #Return the 2nd column. The 2nd columns is probabiltity of success.
    #Returns an array of 1s and 0s
    return threshold_adjusted_preds[:,1]
thres_preds = predictor(LogisticRegression, X_train, X_test, y_train, .9)
profit_calculator(y_test, thres_preds)
```

```python
thres_preds = predictor(LogisticRegression, X_train, X_test, y_train, .7) #making model more conservative
profit_calculator(y_test, thres_preds) #logestic regression is best working with probility based models, naive base, gredientBosting twisting threshhold.. 
```
